```python
X = df2.drop("stress_level", axis=1)
y = df2["stress_level"]
```

```python
#train test datası ayırma 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=15
)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train_scaled")
plt.show()
```

```python
X.shape
```

```python
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
msa=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("mse:",mse)
print("msa:",msa)
print("r2:",r2)
```

```python
#Anksiyete seviyesi ile stres seviyesi arasında doğrusal bir ilişki var mı?
sns.lmplot(x="anxiety_level", y="stress_level", data=df2, ci=None)
plt.title("Anxiety vs Stress")
plt.show()
```

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df2["anxiety_level"], df2["stress_level"])
print("Correlation:", corr)
print("P-value:", p_value)
```

```python
#Düşük özsaygı depresyonu artırıyor mu?
sns.set(style="whitegrid")

g = sns.lmplot(x="self_esteem", y="depression", data=df2, height=6, aspect=1.2, scatter_kws={'alpha':0.5})
g.set_axis_labels("Self-Esteem", "Depression")
g.fig.suptitle("SELF-ESTEEM vs DEPRESSION", fontsize=14, y=1.03)

plt.show()
```

```python
#Öz saygı
orr,p_value=pearsonr(df2["self_esteem"],df2["depression"])
print("Corr:",corr)
print("p_value:",p_value)
```

```python
#----------Etiketler değiştiirlip farklı analizler yapılabilir---------------
```

```python
X = df2[["self_esteem"]]
y = df2["depression"]

# Modeli eğitme
model = LinearRegression()
model.fit(X, y)

# Regresyon katsayısı ve intercept değerini yazdırma
print("Katsayı:", model.coef_[0])
print("Intercept:", model.intercept_)

# Model tahmini
y_pred = model.predict(X)

# Tahmin sonuçlarını scatter plot üzerinde çizme
plt.scatter(X, y, label="Gerçek Değerler")
plt.plot(X, y_pred, color="red", label="Regresyon Doğrusu")
plt.xlabel("Self-esteem")
plt.ylabel("Depression")
plt.title("Doğrusal Regresyon Sonuçları")
plt.show()
```
